```python
#  2. (b) Check if there are any missing values in the dataset. If there are, describe how you
#  would handle them (drop or fill).
print(df.isnull().sum())     #sum() counts True values only
print(df.shape)
```

```python
df.dropna(inplace=True)
print(df.isnull().sum())
df.head(5)
```

```python
groupby_countries=df.groupby('country')
cases_groupby_countries=groupby_countries['Cases'].sum().sort_values(ascending=False)
cases_groupby_countries
```

```python
top10_countries_by_cases=cases_groupby_countries.head(10)
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x=top10_countries_by_cases.index, y=top10_countries_by_cases.values, palette='viridis')
plt.xlabel("County")
plt.ylabel("Number of Cases")
plt.title("Top 10 Countries with Maximum Cases")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Tests',y='Cases',data=df, hue='Deaths',palette='viridis', color='country')
plt.legend(title='Deaths', loc='upper right')
plt.title('Tests vs Cases')
plt.xlabel('Tests')
plt.ylabel('Cases')
plt.show()
```

```python
fig = px.scatter(df,
                 x='Tests',
                 y='Cases',
                 color='country',  # Color points by 'Deaths'
                 hover_data=['country'], # Show country on hover
                 title='COVID-19 Tests vs. Cases',
                 labels={'Tests': 'Number of Tests', 'Cases': 'Number of Cases', 'Deaths': 'Number of Deaths'},
                 template="plotly_white", # Use a clean template
                 width=800, height=600)


fig.update_traces(marker=dict(size=8, line=dict(width=1, color='DarkSlateGrey')),
                  selector=dict(mode='markers')) # Adjust marker appearance

fig.update_layout(xaxis_title="Number of Tests", yaxis_title="Number of Cases",
                   title_font_size=20, font_size=12)

fig.show()
```
