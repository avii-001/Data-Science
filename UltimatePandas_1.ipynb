```python
# Some more analysis
# How many developers code as a hobby, and how does it vary across different countries? Hint: Hobbyist and Country
# What is the average number of years developers have been coding professionally? Hint: YearsCodePro
# How does job satisfaction vary based on the number of working hours per week? Hint: JobSat and WorkWeekHrs
# What are the most common web frameworks used by developers, and are there any preferences for the next year? Hint: WebFrameWorkedWith and WebFrameDesireNextYear
# Is there any relationship between compensation and the number of years of coding experience? Hint: ConvertedComp
# Do developers who work remotely report higher career satisfaction compared to those who don't? Hint: WorkRemote and JobSat
# What is the average age of developers across different countries, and how does it affect their career satisfaction? Hint: CareerSat
```

```python
# How many developers code as a hobby, and how does it vary across different countries? Hint: Hobbyist and Country
df.groupby('Country')['Hobbyist'].value_counts().unstack()
```

```python
# What is the average number of years developers have been coding professionally? Hint: YearsCodePro
df['YearsCodePro']=df['YearsCodePro'].replace('More than 50 years','51')
df['YearsCodePro']=df['YearsCodePro'].replace('Less than 1 year','0')
df['YearsCodePro'].value_counts()
df['YearsCodePro']=df['YearsCodePro'].astype(float)
avg_years=df['YearsCodePro'].mean()
print(f'The average number of years developers have been coding professionally is {avg_years:.2f} yrs')
```

```python
# How does job satisfaction vary based on the number of working hours per week? Hint: JobSat and WorkWeekHrs
df.groupby('JobSat')['WorkWeekHrs'].agg(['mean','median','max'])
```

```python
# What are the most common web frameworks used by developers, and are there any preferences for the next year?
# Hint: WebFrameWorkedWith and WebFrameDesireNextYear
web_frameworks_used=df['WebFrameWorkedWith'].str.split(';',expand=True).stack().value_counts()
web_frameworks_used
```

```python
web_frame_desired_next_year=df['WebFrameDesireNextYear'].str.split(';',expand=True).stack().value_counts()
web_frame_desired_next_year
```

```python
# Is there any relationship between compensation and the number of years of coding experience? Hint: ConvertedComp
df['YearsCode']=df['YearsCode'].replace('Less than 1 year','0')
df['YearsCode']=df['YearsCode'].replace('More than 50 years','51')
df['YearsCode']=df['YearsCode'].astype(float)
df.groupby('YearsCode')['ConvertedComp'].agg(['mean','median']).sort_values(by='YearsCode', ascending=True)
```

```python
# Do developers who work remotely report higher career satisfaction compared to those who don't? Hint: WorkRemote and JobSat
df.groupby('WorkRemote')['JobSat'].value_counts().unstack()           #unstack for readibility & stack for easy graph
```

```python
# What is the average age of developers across different countries, and how does it affect their career satisfaction? Hint: CareerSat
df_avg_age=df.groupby('Country')['Age'].agg('mean')
df_careersat_by_country=df.groupby('Country')['CareerSat'].value_counts().unstack()
print(df_avg_age)
df_careersat_by_country
```

```python
# Task: Sort a list of tuples based on the second element using a lambda function.

# Instructions:

# Create a list of tuples:
# text

# Copy
students = [("Alice", 25), ("Bob", 20), ("Charlie", 30)]
# .
# Use the sorted() function with a lambda to sort the list by the second element (age).
# Print the sorted list.

sorted_students=sorted((student,num) for student, num in students)
sorted_students
```
