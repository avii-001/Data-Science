```python
import pandas as pd
import numpy as np
```

```python
# 1. Load and Explore the Dataset
# • Load the dataset into a Pandas DataFrame.
df=pd.read_csv('/content/drive/MyDrive/kusumm/Data Science/data/titanic.csv')
```

1. Load and Explore the Dataset


```python
# • Display the first 10 rows of the dataset.
df.head(10)
```

```python
# • Print the column names, number of rows and columns, and data types of each column.
print(f'Columns in Dataset\n {df.columns}')
print(f'\nRows & Columns in Dataset\n {df.shape}')
print(f'\nData types in Datasets\n{df.dtypes} ')
```

```python
# • Check for missing values and count them for each column.
df.isnull().sum()
```

2. Data Selec on and Basic Opera ons

```python
# • Select the Name, Age, and Survived columns and display the first 5 rows.
df[['Name','Age','Survived']].head(5)
```

```python
# Filter and display the rows where:
# o The passenger is a female (Sex == 'female') and survived (Survived == 1).
filter_1survived_female=df[(df['Sex']=='female') & (df['Survived']==1)]
filter_1survived_female
```

```python
# o The passenger is younger than 18 years old (Age < 18).
df[df['Age']<18]
```

```python
# • Create a new column called FamilySize that is the sum of SibSp and Parch.
df['FamilySize']=df['SibSp']+df['Parch']                                    # Adding two rows
df[['SibSp','Parch','FamilySize']]
```

3. Data Filtering with loc


```python
# • Use the loc func on to:
# o Select and display passengers who are in Pclass 1 and survived.
Pclass1_survived=(df['Pclass']==1) & (df['Survived']==1)
df.loc[Pclass1_survived]                                                    # .loc gives rows that satisfies the condition
```

```python
# o Update the Age of passengers with missing values to the median age of the dataset.
df['Age'].fillna(df['Age'].median(),inplace=True)
df['Age'].isnull().sum()
```

```python
# o Display rows where the Fare is greater than the 75th percen le of the Fare column.
```

```python
df.describe()
fare_75p=df['Fare'].describe()['75%']                                             #using descibe function to get 75 percentile of 'Fare'
print(f'70 percentile of Fare= {fare_75p}\n')
df[df['Fare']>(df['Fare'].describe()['75%'])]
```

4. Apply Func on

```python
# • Write a custom func on that:
# o Categorizes passengers as Child if their age is less than 18, Adult if their age is between 18 and 60, and Senior if their age is greater than 60.
categorize_age=lambda age: 'Child' if age<18 else ('Adult' if age<60 else 'Senior')
```

```python
# • Apply this func on to the Age column and create a new column called AgeCategory.
df['AgeCategory']=df['Age'].apply(categorize_age)
df[['Age','AgeCategory']]
```

5. Aggregate on and Insights


```python
# • Calculate the average Fare for each Pclass and Embarked group.
df.groupby(['Pclass','Embarked'])['Fare'].mean().round(2).rename('AvgFare')           #Also rounding off to 2 decimals & renaming title to AvgFare
```

```python
# • Count the number of survivors for each Pclass.
df['Pclass'].value_counts().rename('SurvivorsNo.')
# OR
survived_by_pclass=df.groupby('Pclass')['Survived'].sum()
survived_by_pclass
```

```python
# • Find the maximum and minimum age of passengers who survived.
df.groupby('Survived')['Age'].agg(['max','min'])                      # using agg function to get max & min Age of Survived
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# • Create a bar chart showing the survival count for each Pclass.
survived_by_pclass_df=pd.DataFrame(survived_by_pclass)             # converting into Dataframe(sns only uses df)
plt.figure(figsize=(8,4))                                          #figure size
sns.barplot(
    data=survived_by_pclass_df,                                    #data used = survived_by_pclass_df
    x='Pclass',
    y='Survived',
    palette='magma'                                                 # color palatte used (Some most attractives: coolwarm, Blues, crest)
)
plt.title('Survival Count by Pclass')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Count')
plt.show()                                                          # Display graph

passengers_by_pclass=df.groupby('Pclass')['PassengerId'].count()
print(passengers_by_pclass)
```
